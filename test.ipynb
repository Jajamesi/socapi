```python
active_ids_20 = [list(active_polls.keys())[i] for i in range(20)]
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import os

for poll_id in active_ids_20:
    blocks = await session.get_blocks(poll_id=poll_id)
    last_block = await session.find_last_item(blocks)
    block_questions = await session.get_questions(parent_id=last_block.id, how="block")
    last_question = await session.find_last_item(block_questions, question_types={1,2,3,4,8,9,10,11})
    soup = BeautifulSoup(last_question.title, 'html.parser')
    text = soup.get_text(strip=True)

    await session.download(poll_id=poll_id, export_format="xlsx")

    if os.path.exists(f"poll_{poll_id}.xlsx"):
        df = pd.read_excel(f"poll_{poll_id}.xlsx")

        print(poll_id, " - ", text in df.columns)


        # last_df_col = df.columns[-1]

        # print(poll_id, " - ", last_df_col == text)
```
