```python
import asyncio

async def print_with_delay(x):
    await asyncio.sleep(x % 5)  # Simulate a delay up to 5 seconds
    print(f"Task {x} completed with delay {x % 5} seconds")

async def main():
    for i in range(40):
        # Launch the task and let it run concurrently
        asyncio.create_task(print_with_delay(i))
        await asyncio.sleep(0.1)  # Simulate some processing time in the loop

# Run the event loop
if __name__ == "__main__":
    asyncio.run(main())
```
